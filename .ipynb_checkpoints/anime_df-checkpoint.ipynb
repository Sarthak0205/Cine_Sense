```python
anime_df.columns = anime_df.columns.str.lower().str.strip()
anime_df.columns
```

```python
anime_df = anime_df[['title', 'synopsis', 'airing_status', 'mpaa_rating']]
anime_df.head()
```

```python
anime_df = anime_df.dropna(subset=['synopsis'])

anime_df['airing_status'] = anime_df['airing_status'].fillna('')
anime_df['mpaa_rating'] = anime_df['mpaa_rating'].fillna('')
```

```python
anime_df['synopsis'] = anime_df['synopsis'].apply(lambda x: x.lower())
anime_df['title'] = anime_df['title'].apply(lambda x: x.lower())
```

```python
anime_df['synopsis'] = anime_df['synopsis'].astype(str)
anime_df['title'] = anime_df['title'].astype(str)
anime_df['airing_status'] = anime_df['airing_status'].astype(str)
anime_df['mpaa_rating'] = anime_df['mpaa_rating'].astype(str)
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)
    return text

anime_df['synopsis'] = anime_df['synopsis'].apply(clean_text)
```

```python
anime_df['tags'] = (
    anime_df['title'] + " " +
    anime_df['title'] + " " +   # repeat = boost importance
    anime_df['synopsis']
)
```

```python
anime_df = anime_df[['title', 'tags']]
anime_df.head()
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
vectors_anime = model.encode(anime_df['tags'].tolist(), show_progress_bar=True)

from sklearn.metrics.pairwise import cosine_similarity
similarity_anime = cosine_similarity(vectors_anime)
```

```python
def recommend_anime(title):
    title = title.lower()
    
    matches = anime_df[anime_df['title'].str.contains(title, case=False)]

    if matches.empty:
        print("Anime not found")
        return

    idx = matches.index[0]
    
    idx = anime_df[anime_df['title'] == title].index[0]
    distances = similarity_anime[idx]
    
    anime_list = list(enumerate(distances))
    
    anime_list = sorted(
        anime_list,
        reverse=True,
        key=lambda x: x[1]
    )[1:6]
    anime_list = [i for i in anime_list if "robot" not in anime_df.iloc[i[0]].tags][:5]
    for i in anime_list:
        print(anime_df.iloc[i[0]].title)
```

```python
recommend_anime("naruto")
```
